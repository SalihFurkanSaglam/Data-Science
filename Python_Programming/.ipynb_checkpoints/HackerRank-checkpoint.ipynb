## **If-Else**


```python
n = int(input())

if (n%2==0 and (2<n<=5 or n>20)):
    print("Not Weird")
else:
    print("Weird")
```

```python
x = int(input()) 
y = int(input()) 
z = int(input()) 
n = int(input()) 
print([[i, j, k] for i in range(x+1) for j in range (y+1) for k in range (z+1) if ((i+j+k) != n)])
```
